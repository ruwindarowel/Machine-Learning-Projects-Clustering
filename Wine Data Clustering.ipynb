## K Means

```python
wss={}

for i in range(1,10):
    kmeans=KMeans(n_clusters=i,max_iter=1000)
    kmeans.fit(scaled_df)
    wss[i]=kmeans.inertia_

sns.lineplot(x=wss.keys(),y=wss.values(),markers=True)
#sns.barplot(x=silhoutte_ss.keys(),y=silhoutte_ss.values())
plt.xlabel("Values of 'k'")
plt.ylabel('WCSS')
plt.show()
```

```python
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(scaled_df, KMeans(n_clusters = n_cluster).fit_predict(scaled_df))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
print("..............................................................")
ss_df=pd.DataFrame({"Cluster":k,"Silhoutte_Score":silhouette_scores})
print(ss_df)
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(scaled_df)
```
